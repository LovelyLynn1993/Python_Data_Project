```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_filtered = df
```

```python
cleaned = df_filtered.dropna(subset=['salary_year_avg'])
rows_after = len(cleaned)
rows_before = len(df_filtered)

rows_before, rows_after
```

```python
update_df = df_filtered.drop_duplicates(subset=['job_location'])
rows_after = len(update_df)
rows_before = len(df_filtered)

rows_before, rows_after
```

```python
rate_before = df_filtered['salary_rate'].head(10)
df_filled = df_filtered.fillna({'salary_rate': 'Unknown'})
rate_after = df_filled['salary_rate'].head(10)

rate_before, rate_after
```
