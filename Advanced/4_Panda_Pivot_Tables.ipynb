```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_copy = df.copy()

id(df_copy) == id(df)
```

```python
df_table = df_copy.pivot_table(index=['job_title_short', 'job_country'], aggfunc='size')

df_table.head(10)
```

```python
df_alt = df.copy()
id(df_alt) == id(df)
```

```python
pv_table = df_alt.pivot_table(values='salary_year_avg', index='company_name', aggfunc=['max', 'min', 'mean'])

pv_table.head(5)
```

```python
df_pv = df.copy()
id(df_pv) == id(df)
```

```python
df_filtered = df_pv[df_pv['job_title_short'] == 'Data Scientist'].copy()
piviot_table = df_filtered.pivot_table(values='salary_year_avg', index=['company_name', 'job_country'], aggfunc= 'median')
alt_piviot_table = piviot_table[piviot_table['salary_year_avg'] > 200000]

alt_piviot_table.head(10)
```
