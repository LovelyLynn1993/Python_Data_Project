```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_us_software_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col='job_posted_month')

df_us_software_pivot
```

```python
df_us = df[df['job_country'] == 'United States'].copy()
df_us['job_posted_month'] = df_us['job_posted_date'].dt.strftime('%B')
df_us_pivot = df_us.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_us_pivot = df_us_pivot.reset_index()
df_us_pivot['job_posted_month_no'] = pd.to_datetime(df_us_pivot['job_posted_month'], format='%B').dt.month
df_us_pivot = df_us_pivot.sort_values('job_posted_month_no')
df_us_pivot = df_us_pivot.set_index('job_posted_month')
df_us_pivot = df_us_pivot.drop(columns='job_posted_month_no')

df_us_pivot 
```

```python
df_us_companies = df_us_software_pivot.merge(df_us_pivot, how='right', on='job_posted_month' )
df_us_companies.head()
```

```python
df_us_jobs = df[df['job_country'] == 'United States'].copy()
df['job_posted_month'] = df['job_posted_date'].dt.to_period('M')
df_monthly_postings = df_us_jobs.groupby('job_posted_month').size().reset_index(name='postings_count')
merged_df = pd.merge(df_us_jobs, df_monthly_postings, on='job_posted_month', how='right')

merged_df.head()
```

```python
df_not_remote_jobs = df.copy()
```

```python
df_not_remote_jobs = df[df['job_work_from_home'] == False]
df_not_remote_salary = df_not_remote_jobs.groupby('job_title_short')['salary_year_avg'].mean().reset_index()
df_not_remote_salary.rename(columns={'salary_year_avg': 'salary_not_remote_avg'}, inplace=True)
merge_df = pd.merge(df_not_remote_jobs, df_not_remote_salary, on='job_title_short', how='right')

merge_df.head()
```

```python
df_extra = df.copy()
```

```python
df_company_salary = df_extra.groupby('company_name')['salary_year_avg'].mean().reset_index()
df_company_remote = df_extra[df_extra['job_work_from_home'] == True].groupby('company_name').size().reset_index(name='remote_count')
name_merge = pd.merge(df_company_salary, df_company_remote, on='company_name', how='inner')


name_merge.head()
```
