```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_alt = df.copy()
id(df_alt) == id(df)
```

```python
data_scientist_df = df_alt[df_alt['job_title'] == 'Data Scientist']
data_analyst_df = df_alt[df_alt['job_title'] == 'Data Analyst']
combined_df = pd.concat([data_scientist_df, data_analyst_df], ignore_index=True)

combined_df
```

```python
df_a = df.copy()
id(df_a) == id(df)
```

```python
df_a['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
```

```python
df_oct = df_a[df_a['job_posted_month'] == 'Oct']
df_nov = df_a[df_a['job_posted_month'] == 'Nov']
df_dec = df_a[df_a['job_posted_month'] == 'Dec']
```

```python
combined_df = pd.concat([df_oct, df_nov, df_dec], ignore_index=True)

combined_df
```

```python
df_c = df.copy()
id(df_c) == id(df)
```

```python
full_time_df = df_c[df_c['job_schedule_type'] == 'Full-time']
part_time_df = df_c[df_c['job_schedule_type'] == 'Part-time']
combined_df = pd.concat([full_time_df, part_time_df], ignore_index=True)

combined_df
```
