```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_date_str'] = df['job_posted_date'].apply(lambda x: x. strftime('%Y-%m-%d'))
df[['job_posted_date', 'job_posted_date_str']].head()
```

```python
import datetime as dt
```

```python
current_date = dt.datetime.now()
df['days_since_posted'] = df['job_posted_date'].apply(lambda x: (current_date - x).days)
df[['job_posted_date', 'days_since_posted']].head()
```

```python
df_filtered = df.dropna(subset=['salary_year_avg']).copy()
```

```python
df_filtered['salary_category'] = df_filtered.apply(lambda row: 'Low' if row['salary_year_avg'] < 60000 else 'Medium' if row['salary_year_avg'] <= 100000 else 'High', axis=1)

df_filtered[['salary_year_avg', 'salary_category']].head()
```
